## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

- **The plot method on Pandas Series and DataFrames is just a simple wrapper around plt.plot():**
- **Define a Pandas Dataframe with all crypto allocation in a given portfolio**
- **Note that total sum = 100%** 
- **Use matplotlib to plot a pie chart** 

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

%matplotlib inline

import warnings

warnings.filterwarnings("ignore") 
```

```python
crypto = pd.read_csv("crypto_daily_prices.csv")
```

```python
crypto
```

```python
crypto["Date"] = pd.to_datetime(crypto["Date"])
crypto
```

```python
crypto.set_index("Date",inplace=True)
crypto
```

```python
crypto['allocations_total'] = crypto.iloc[:,-3:-1].sum(axis=1)
crypto.sample(5)
```

```python
crypto['Percentage_BTC'] = (crypto["BTC"] / crypto["allocations_total"]) * 100
crypto['Percentage_ETH'] = (crypto["ETH"] / crypto["allocations_total"]) * 100
crypto['Percentage_ADA'] = (crypto["ADA"] / crypto["allocations_total"]) * 100

crypto.sample(5)
```

```python
crypto_tail = crypto.tail(1) 

crypto_alloc = crypto_tail[['Percentage_BTC', 'Percentage_ETH', 'Percentage_ADA']]

crypto_alloc
```

```python
crypto_alloc.rename(columns={"Percentage_BTC": "BTC", "Percentage_ETH": "ETH", "Percentage_ADA": "ADA"}, inplace = True) 
crypto_alloc
```

```python
my_values = crypto_alloc.values.round()
my_values 
```

```python
my_labels = crypto_alloc.columns
my_labels 
```

```python
plt.figure(figsize=(12,8))

my_values = [94.931608, 5.068392, 0.002261]
plt.pie(my_values, labels=my_labels, autopct='%1.1f%%', startangle=140, explode = [0.6, 0.2, 0.2])

plt.axis('equal')
plt.show()
```

## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
- **A histogram represents data using bars with various heights**
- **Each bar groups numbers into specific ranges**
- **Taller bars show that more data falls within that specific range**
- **Plot the histogram of the returns**

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
num_bins = 40

plt.figure(figsize=(12, 8))
sns.histplot(daily_return_df['BTC'].dropna(), bins=num_bins, edgecolor='black', color='orange',kde=True)

plt.xlabel('BTC Daily Return')
plt.ylabel('Density')

plt.show()
```

```python
mu = round(daily_return_df['ETH'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['ETH'].std(), 2) # standard deviation of distribution
num_bins = 40

plt.figure(figsize=(12, 8))
sns.histplot(daily_return_df['ETH'].dropna(), bins=num_bins, edgecolor='black', color='red',kde=True)

plt.xlabel('ETH Daily Return')
plt.ylabel('Density')

plt.show()
```

```python
mu = round(daily_return_df['ADA'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['ADA'].std(), 2) # standard deviation of distribution
num_bins = 40

plt.figure(figsize=(12, 8))
sns.histplot(daily_return_df['ADA'].dropna(), bins=num_bins, edgecolor='black', color='brown',kde=True)

plt.xlabel('ADA Daily Return')
plt.ylabel('Density')

plt.show()
```

```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='mean area', y='mean smoothness', data=cancer_df,color="red")

plt.title('Scatter Plot Between Mean Area and Mean Smoothness')
plt.xlabel('Mean Area')
plt.ylabel('Mean Smoothness')

plt.show()
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
plt.figure(figsize=(12,8))
sns.pairplot(cancer_df, vars=['mean radius', 'mean perimeter', 'mean area', 'mean smoothness'], kind='hist')
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr_matrix = cancer_df[['mean radius', 'mean perimeter', 'mean area', 'mean perimeter']].corr()
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True, cmap='crest')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(cancer_df['mean radius'], kde=True, hist=True, color='blue')

plt.title('Distribution of Mean Radius')
plt.xlabel('Mean Radius')
plt.ylabel('Probability Density')
plt.show()
```

PRACTICE OPPORTUNITY 
- **Plot two separate distplot for each target class #0 and target class #1**


```python

class_0_data = cancer_df[cancer_df['target'] == 0]
class_1_data = cancer_df[cancer_df['target'] == 1]

plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
sns.distplot(class_0_data['mean radius'], kde=True, hist=True, color='blue')
plt.title('Class 0')
plt.xlabel('Mean Radius')
plt.ylabel('Probability Density')


plt.subplot(1, 2, 2)
sns.distplot(class_1_data['mean radius'], kde=True, hist=True, color='red')
plt.title('Class 1')
plt.xlabel('Mean Radius')
plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```
